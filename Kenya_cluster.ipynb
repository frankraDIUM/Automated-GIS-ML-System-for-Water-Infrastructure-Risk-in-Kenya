```python
# Visualise Data in points and centroids

import leafmap
import geopandas as gpd
import matplotlib.pyplot as plt

# Load files (adjust paths)
points = gpd.read_file("I:/GEO DATA ANALYSIS/Kenya Water/nairobi_water_project/clusters_points.geojson")

centroids = gpd.read_file("I:/GEO DATA ANALYSIS/Kenya Water/nairobi_water_project/clusters_centroids.geojson")

# Create map centered on Kenya
m = leafmap.Map(center=[0.5, 36.8], zoom=6)

# Add clustered points (color by cluster_id)
m.add_gdf(
    points,
    layer_name="Non-functional points",
    style_callback=lambda feature: {
        "color": f"hsl({feature['properties']['cluster_id'] * 30}, 80%, 50%)",
        "radius": 3
    }
)

# Add centroids (size by cluster size)
m.add_gdf(
    centroids,
    layer_name="Cluster Centroids",
    style_callback=lambda feature: {
        "color": "black",
        "fillColor": "red",
        "radius": min(feature['properties']['size'] / 10, 20),
        "fillOpacity": 0.7
    }
)

m.add_layer_control()
```
